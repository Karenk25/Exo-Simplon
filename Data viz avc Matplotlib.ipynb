```python
import pandas as pd
# pour les graphiques apparaissant dans le notebook
%matplotlib inline
#importation de la librairie
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Téléchargements/chomage_zone.xls', sheet_name="txcho_ze", skiprows=5)
data
```

```python
nom_col = ['REG', 'LIB', 'PE2018']
p_emploi = pd.read_excel('Téléchargements/pole_emploi.xlsx', names=nom_col, sheet_name='Data', skiprows=3)
p_emploi
```

```python
data_grp = data.groupby(['LIBREG'], as_index=False).mean()
data_grp
```

```python
#une taille correct au graphique, lui donner un titre et ajouter une grille en arrière-plan
data_grp.plot(x='LIBREG',y='2020-T1', kind='bar', title='Taux de chômage', grid='default' ,figsize=(20,10))
```

```python
data2=data.iloc[:,3:74]
data2
```

```python
data_grp2 = data2.groupby(['LIBREG']).mean()
data_grp2
```

```python
#Transposer le dataframe.
data_grp_t = data_grp2.transpose()
data_grp_t
```

```python
#réindexer le dataframe avec reset_index().
data_grp_t = data_grp_t.reset_index()
data_grp_t
```

```python
#Renommer le nom de la colonne ‘index’ en ‘periode’.
data_grp_rename = data_grp_t.rename(columns={'index': 'periode'})
data_grp_rename
```

```python
#Après avoir visualisé le contenu de data2, utiliser pandas.plot() pour tracer la ligne
#représentant l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte
#d'Azur’.
data_grp_rename.plot(kind='line', x='periode', y="Provence-Alpes-Côte d'Azur", 
                     title= 'Taux de chomage',grid='default' ,figsize=(20,10))
```

```python
#Répéter l’étape précédente sans préciser l’argument « y = ».
data_grp_rename.plot(kind='line', x='periode', title= 'Taux de chomage',grid='default' ,figsize=(20,10))
```

```python
#Représenter l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte
#d'Azur’, ‘Occitanie’ et ‘Ile-de-France’ sur un même graphique.

data_grp_rename.plot(kind='line', x='periode', y=["Provence-Alpes-Côte d'Azur","Occitanie","Ile-de-France"],
                     title= 'Taux de chomage',grid='default' ,figsize=(20,10))
```
